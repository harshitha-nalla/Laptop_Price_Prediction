```python
def processor_type(text):
    if text == 'Intel Core i5' or text == 'Intel Core i3' or text == 'Intel Core i7':
        return text
    else:
        if text.split()[0] == 'Intel':
            return 'Other Intel Processor'
        else:
            return 'AMD Processor'
```

```python
data['Memory'] = data['Memory'].astype(str).replace(r'\.0', '', regex=True)
def changeTBtoGB(text):
    wordArray = text.split(' ')
    for i in range(len(wordArray)):
        if 'TB' in wordArray[i]:
            word = wordArray[i].strip()
            word = word[:-2]
            word = int(word)*1024
            wordArray[i] = f'{word}'
    return " ".join(wordArray)
data['Memory'] = data['Memory'].apply(changeTBtoGB)
data['Memory'] = data['Memory'].str.replace('GB', '')
sizeOfdata = len(data)
index = len(data.columns)-1
newValues = {
    'SSD': [0]*sizeOfdata,
    'HDD': [0]*sizeOfdata,
    'Flash_Storage': [0]*sizeOfdata,
    'Hybrid': [0]*sizeOfdata
}

for col_name, col_values in newValues.items():
    data.insert(index, col_name, col_values)
    index+=1

newdata = data['Memory'].str.split('+', n=1, expand=True)
newdata[1] = newdata[1].fillna('0')
newdata[0] = newdata[0].str.strip()
newdata[1] = newdata[1].str.strip()
first = newdata[0].str.findall(r'^([0-9]{0,4}.d*)').apply(lambda x: int(x[0]))
second = newdata[1].str.findall(r'^([0-9]{0,4}.d*)').apply(lambda x: int(x[0]))
for index, row in newdata.iterrows():
    if 'SSD' in row[0]:
        data.loc[index, 'SSD'] += first.loc[index]
    if 'HDD' in row[0]:
        data.loc[index, 'HDD'] += first.loc[index]
    if 'Flash Storage' in row[0]:
        data.loc[index, 'Flash_Storage'] += first.loc[index]
    if 'Hybrid' in row[0]:
        data.loc[index, 'Hybrid'] += first.loc[index]
    if 'SSD' in row[1]:
        data.loc[index, 'SSD'] += second.loc[index]
    if 'HDD' in row[1]:
        data.loc[index, 'HDD'] += second.loc[index]
    if 'Flash Storage' in row[1]:
        data.loc[index, 'Flash_Storage'] += second.loc[index]
    if 'Hybrid' in row[1]:
        data.loc[index, 'Hybrid'] += second.loc[index]
data.head(50)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
```

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor
from sklearn.metrics import r2_score, mean_absolute_error
```

```python
s1=ColumnTransformer(transformers=[('col',OneHotEncoder(sparse_output=False,drop='first',handle_unknown='ignore'),[0,1,9,10,11])],remainder='passthrough')
s2=LinearRegression()
pipe=Pipeline([('s1',s1),('s2',s2)])
pipe.fit(X_train,y_train)
y_predict=pipe.predict(X_test)
print('R2 Score',r2_score(y_test,y_predict))
print('MAE',mean_absolute_error(y_test,y_predict))
```
